```python
import numpy as np
```

```python
#Creating Numpy array with zeros

#1D
arr_zero=np.zeros(4)
print(arr_zero)
print(" ")

#More than 1D
arr_zero1=np.zeros((3,4))
print(arr_zero1)
```

```python
#Creating Numoy array with ones

#1D
arr_one=np.ones(4)
print(arr_one)
print(" ")

#More than 1D
arr_one1=np.ones((4,3))
print(arr_one1)
```

```python
#Creating Empty Array
arr_empt=np.empty(4)
print(arr_empt)
```

```python
#Array Diagonal elements filled with ones
arr_diag=np.eye(4)
print(arr_diag)

#if not a square matrix
arr_diag1=np.eye(3,7)
print(arr_diag1)
```

```python
#Creating a array with elements in it with equally spaced
arr_spaced=np.linspace(0,20,5)
print(arr_spaced)
```
