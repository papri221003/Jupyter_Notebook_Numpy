```python
import numpy as np
```

```python
#Creating Numpy array with zeros

#1D
arr_zero=np.zeros(4)
print(arr_zero)
print(" ")

#More than 1D
arr_zero1=np.zeros((3,4))
print(arr_zero1)
```

```python
#Creating Numoy array with ones

#1D
arr_one=np.ones(4)
print(arr_one)
print(" ")

#More than 1D
arr_one1=np.ones((4,3))
print(arr_one1)
```

```python
#Creating Empty Array
arr_empt=np.empty(4)
print(arr_empt)
```
